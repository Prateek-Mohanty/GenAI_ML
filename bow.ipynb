```python
import pandas as pd
import nltk

messages = pd.read_csv('spamhamdata.csv',sep="\t",names=["label","message"])
```

```python
messages
```

```python
import re
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
from nltk.stem import SnowballStemmer

sst = SnowballStemmer('english')
```

```python
corpus = []

for i in range (0,len(messages)):
    review = re.sub("[^a-zA-Z]", " ", messages['message'][i])
    review = review.lower()
    review = review.split()
    review = [sst.stem(word)for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
for i in range (len(corpus)):
    print(f"This is my {i} sentence : {corpus[i]}")
```

```python
## Lets go and create the BOW model
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500, binary=True,vocabulary=None)
```

```python
X=cv.fit_transform(corpus).toarray()
```
