```python
from nltk.stem import PorterStemmer
from nltk.stem import SnowballStemmer
from nltk.stem import WordNetLemmatizer
```

```python
words = ["going","goes","drowning","history"]
```

```python
porter = PorterStemmer()

for word in words:
    print(f"porter {word} ----> {porter.stem(word)}")
```

```python
snowball = SnowballStemmer(language="english")

for word in words:
    print(f"snowball {word} ----> {snowball.stem(word)}")
```

```python
wordnet = WordNetLemmatizer()

for word in words:
    print(f"lemma {word} ----> {wordnet.lemmatize(word,pos='v')}")
```
